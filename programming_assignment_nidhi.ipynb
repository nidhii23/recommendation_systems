`explaination`
`user ratings` filters the ratings dataframe to get only the ratings from the specified user (user_id).
`user movie_matrix` Creates a matrix where rows represent users, columns represent movies, and values are ratings.
uses 0 to fill in missing values (unrated films receive a 0).
` User similarity ` and `user sim dif`calculates each user's cosine similarity, which measures how similar their rating behaviours are.
 For simpler reference, the similarity matrix is wrapped in a DataFrame with user IDs as row and column names.
`similar user` selects the target user's column from the similarity matrix.
 people are sorted by similarity score in descending order, with the most similar people appearing first.
 Users are taken from positions 1 through 5 (position 0 is skipped since it is the user with a perfect similarity of 1.0).
`user rated movies` `similar users movies`  ` candidate movies`  generates a collection of movie IDs that have already been rated by the intended user.
 obtains all ratings from the previously determined similar users.
 Only ratings for films that the target user hasn't yet reviewed are retained by filters (possible suggestions).
`movie ratings` Sorts relevant films according to movie_id.
 determines a weighted average rating for every film, with the weights being the similarity scores.
 This implies that the final score is influenced more by ratings from users who are more alike.
 Movies are sorted from top to lowest based on their weighted ratings.
`top movies ` Takes the top 'num' movies with the highest weighted ratings.  `movies name ` Translates the movie IDs back to their titles for better readability.
Uses a movies dataframe.

````python

import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

def recommend_movies_for_user(user_id, num=5):
    # Get user's rated movies
    user_ratings = ratings[ratings['user_id'] == user_id]

    # Create user-movie matrix
    user_movie_matrix = ratings.pivot(index='user_id', columns='movie_id', values='rating')

    # Calculate user similarity using cosine similarity
    user_similarity = cosine_similarity(user_movie_matrix.fillna(0))
    user_sim_df = pd.DataFrame(user_similarity, index=user_movie_matrix.index, columns=user_movie_matrix.index)

    # Get similar users
    similar_users = user_sim_df[user_id].sort_values(ascending=False)[1:6]  # Get top 5 similar users

    # Get movies rated by similar users but not by target user
    user_rated_movies = set(user_ratings['movie_id'])
    similar_users_ratings = ratings[ratings['user_id'].isin(similar_users.index)]
    candidate_movies = similar_users_ratings[~similar_users_ratings['movie_id'].isin(user_rated_movies)]

    # Calculate weighted average rating
    movie_ratings = candidate_movies.groupby('movie_id').apply(
        lambda x: np.average(x['rating'], weights=similar_users[x['user_id']])
    ).sort_values(ascending=False)

    # Get top N movies
    top_movies = movie_ratings.head(num)

    # Map movie IDs to titles
    movie_names = movies[movies['movie_id'].isin(top_movies.index)]['title'].values

    # Create result DataFrame
    result_df = pd.DataFrame({
        'Ranking': range(1, num+1),
        'Movie Name': movie_names
    })
    result_df.set_index('Ranking', inplace=True)

    return result_df

recommend_movies_for_user(10, num = 5)
````

By reversing the user-movie matrix, item_similarity computes the cosine similarity between movies (not users).  Based on user rating trends, this gauges how similar films are; films that are rated similarly by the same users score higher on similarity.


 For easy reference, `item_sim_df` encapsulates the similarity matrix in a DataFrame with movie IDs as row and column names.

 The movie title's unique identifier is retrieved from the movies DataFrame using the movie_id function.  This enables the machine to search for similarity scores related to this particular film.

 The `sim_scores` function chooses the target movie's column from the similarity matrix and arranges the films in descending order of similarity score (more similar first).

 With a perfect similarity score of <sim_scores = sim_scores.drop(movie_id)>, the movie is eliminated from the list of recommended films.

The top 'num' movie IDs that have the highest similarity scores to the target film are taken by top_movie_ids.
 By restricting the movies DataFrame to only include movies with IDs in the top_movie_ids list, recommended_movies converts the movie IDs back to their titles.
 For improved presentation and readability, result_df formats a DataFrame with movie titles and rankings.


````python

import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
# Calculate item similarity (using cosine similarity on the transposed matrix)
item_similarity = cosine_similarity(user_movie_matrix.T.fillna(0))
item_sim_df = pd.DataFrame(item_similarity, index=user_movie_matrix.columns, columns=user_movie_matrix.columns)

def recommend_movies(movie_name, num=5):
    """Recommends movies similar to a given movie."""
    # Get the movie ID
    movie_id = movies[movies['title'] == movie_name]['movie_id'].iloc[0]

    # Get similarity scores for the movie and sort
    sim_scores = item_sim_df[movie_id].sort_values(ascending=False) # Make sure item_sim_df is defined.

    # Exclude the movie itself
    sim_scores = sim_scores.drop(movie_id)

    # Get top similar movie IDs
    top_movie_ids = sim_scores.head(num).index

    # Get movie titles
    recommended_movies = movies[movies['movie_id'].isin(top_movie_ids)]['title'].tolist()

    # Create result DataFrame
    result_df = pd.DataFrame({'Ranking': range(1, num + 1), 'Movie Name': recommended_movies})
    result_df.set_index('Ranking', inplace=True)

    return result_df

# Example usage
recommendations = recommend_movies("Jurassic Park (1993)", num=5)
print(recommendations)
````

## **Part 3: Graph-Based Recommender (Pixie-Inspired Algorithm)**

### **Adjacency List**

#### **Objective**
In this task, you will preprocess the Movie dataset and construct a **graph representation** where:
- **Users** are connected to the movies they have rated.
- **Movies** are connected to users who have rated them.
  
This graph structure will help in exploring **user-movie relationships** for recommendations.

#### **Step 1: Merge Ratings with Movie Titles**
Since we have **movie IDs** in the ratings dataset but need human-readable movie titles, we will:
1. Merge the `ratings` DataFrame with the `movies` DataFrame using the `'movie_id'` column.
2. This allows each rating to be associated with a **movie title**.

#### **Hint:**
Use the following Pandas operation to merge:
```python

```


#### **Step 2: Aggregate Ratings**
Since multiple users may rate the same movie multiple times, we:
1. Group the dataset by `['user_id', 'movie_id', 'title']`.
2. Compute the **mean rating** for each movie by each user.
3. Reset the index to ensure we maintain a clean DataFrame structure.

#### **Hint:**  
Use `groupby()` and `mean()` as follows:
```python

```

#### **Step 3: Normalize Ratings**
Since different users have different rating biases, we normalize ratings by:
1. **Computing each user's mean rating**.
2. **Subtracting the mean rating** from each individual rating.

#### **Instructions:**
- Use `groupby('user_id')` to group ratings by users.
- Apply `transform(lambda x: x - x.mean())` to adjust ratings.

#### **Hint:**  
Normalize ratings using:
```python
ratings['rating'] = ratings.groupby('user_id')['rating'].transform(lambda x: x - x.mean())
```
This ensures each user’s ratings are centered around zero, making similarity calculations fairer.

#### **Step 4: Construct the Graph Representation**
We represent the user-movie interactions as an **undirected graph** using an **adjacency list**:
- Each **user** is a node connected to movies they rated.
- Each **movie** is a node connected to users who rated it.

#### **Graph Construction Steps:**
1. Initialize an empty dictionary `graph = {}`.
2. Iterate through the **ratings dataset**.
3. For each `user_id` and `movie_id` pair:
   - Add the movie to the user’s set of connections.
   - Add the user to the movie’s set of connections.

#### **Hint:**  
The following code builds the graph:

```python
graph = {}
for _, row in ratings.iterrows():
    user, movie = row['user_id'], row['movie_id']
    if user not in graph:
        graph[user] = set()
    if movie not in graph:
        graph[movie] = set()
    graph[user].add(movie)
    graph[movie].add(user)
```

This results in a **bipartite graph**, where:
- **Users** are connected to multiple movies.
- **Movies** are connected to multiple users.

#### **Step 5: Understanding the Graph**
- **Nodes** in the graph represent **users and movies**.
- **Edges** exist between a user and a movie **if the user has rated the movie**.
- This structure allows us to find **users with similar movie tastes** and **movies frequently watched together**.

#### **Exploring the Graph**
- **Find a user’s rated movies:**  
  ```python
  user_id = 1
  print(graph[user_id])  # Movies rated by user 1
  ```

- **Find users who rated a movie:**  
  ```python
  movie_id = 50
  print(graph[movie_id])  # Users who rated movie 50
  ```


`ratings = ratings.merge(movies, on='movie_id')` joins the ratings DataFrame with the movies DataFrame using movie_id as the common key.
`ratings = ratings.groupby(['user_id', 'movie_id', 'title'])['rating'].mean().reset_index()` calculates the average rating for each user-movie combination if there are duplicate ratings.

`ratings['rating'] = ratings.groupby('user_id')['rating'].transform(lambda x: x - x.mean())` normalizes ratings by subtracting each user's mean rating from their individual ratings.

`graph = {}` creates an empty dictionary to store the bipartite graph representation

The loop through `ratings.iterrows()` builds the graph by:
 Creating empty sets for new users and movies encountered
 Adding bidirectional connections between users and movies. Using sets to ensure unique connections


````python
# Step 1: Merge Ratings with Movie Titles
ratings = ratings.merge(movies, on='movie_id')  # Perform merge before grouping

# Step 2: Aggregate Ratings
ratings = ratings.groupby(['user_id', 'movie_id', 'title'])['rating'].mean().reset_index()
# Step 3: Normalize Ratings
ratings['rating'] = ratings.groupby('user_id')['rating'].transform(lambda x: x - x.mean())
# Step 4: Construct the Graph Representation
graph = {}
for _, row in ratings.iterrows():
    user, movie = row['user_id'], row['movie_id']
    if user not in graph:
        graph[user] = set()
    if movie not in graph:
        graph[movie] = set()
    graph[user].add(movie)
    graph[movie].add(user)

# Step 5
user_id = 1
print(f"Movies rated by user {user_id}: {graph[user_id]}")

movie_id = 50
print(f"Users who rated movie {movie_id}: {graph[movie_id]}")
````

#### **Step 2: Implement the Random Walk Algorithm**

##### **Hints for Implementation**

The weighted_pixie_recommend function performs random walks on a user-movie bipartite graph to generate movie recommendations. It can start from either a user ID or a movie title and returns a list of recommended movies.In this recommendation system, we use a bipartite graph to model the relationships between users and movies. This means we have two separate groups of nodes: one for users and one for movies. A user is connected to a movie if they have rated it, and the strength of that connection is based on the rating — higher ratings create stronger connections, which helps prioritize better-rated movies.

To make the system efficient, we don’t compute the connections on the fly. Instead, we build two dictionaries in advance. One dictionary maps each user to the movies they’ve rated, and the other maps each movie to the users who have rated it. These act like adjacency lists and allow us to quickly look up connections in the graph without needing to recalculate them.

When generating recommendations, we perform weighted random walks on this graph. That means when moving from one node to another — like from a user to a movie, or from a movie to another user — the algorithm chooses the next step based on the rating. The higher the rating, the more likely it is that the algorithm will follow that path. This way, we naturally give more attention to highly rated movies during the recommendation process.
he algorithm can start from either a movie or a user, depending on what input you give.

If you start with a movie title, the algorithm first looks up the movie’s ID based on its name. Once it has the ID, it uses that as the starting point in the graph.

If you start with a user ID, the algorithm uses that number directly, since it's already in the format it needs.

To keep track of where it is during the walk — whether it's currently looking at a user node or a movie node — the algorithm uses a flag called is_user. If is_user is True, it means we’re currently on a user node and need to move to a movie next. If it’s False, we’re on a movie node and should move to a user next. This helps the algorithm alternate properly between users and movies as it walks through the graph. For each user in the system, the algorithm goes through the movies they've rated and focuses on the ones they rated positively — meaning ratings above a certain threshold (like 3 or 4 stars, depending on the design).

Once those positive-rated movies are identified, the algorithm stores them in a dictionary. Each entry in the dictionary corresponds to a user and contains two arrays:

One with the IDs of the movies they liked

One with the actual ratings they gave to those movies

Next, it normalizes those ratings. This means it adjusts the values so that they all add up to 1.0. By doing this, each rating is converted into a probability. So, if a user gave higher ratings to some movies, those movies will have higher probabilities. These probabilities are then used during the random walk to decide which movie the algorithm is most likely to follow from that user.
The recommendation algorithm performs a random walk on the bipartite graph, switching back and forth between users and movies at each step.

When the algorithm is on a user node, it looks at all the movies that user has rated positively. It then chooses one of those movies at random — but not completely randomly. The chances of picking each movie are based on how highly the user rated it. So higher-rated movies are more likely to be chosen.

When the algorithm is on a movie node, it looks at all the users who have rated that movie. Again, it randomly selects one of those users, with the probability based on how highly each user rated that movie.

This back-and-forth process continues for a set number of steps, controlled by a parameter called walk_length.

Even though the walk includes both users and movies, the algorithm only records the movie nodes along the way. This is because the goal is to recommend movies, not users.

To make sure the selections are influenced by the ratings, the algorithm uses np.random.choice with a p parameter — this lets it choose based on the probabilities calculated from the ratings. That way, transitions are weighted, meaning higher-rated items are more likely to be followed during the walk.

````python
import random
import numpy as np
import pandas as pd

def weighted_pixie_recommend(start_point, walk_length=10, num=5):

    # Determine starting point
    if isinstance(start_point, str):
        # Start with a movie title
        movie_row = movies[movies['title'] == start_point]
        if movie_row.empty:
            print("Movie not found.")
            return pd.DataFrame()
        current = movie_row['movie_id'].iloc[0]
        is_user = False  # Current node is a movie
    else:
        # Start with a user ID
        if start_point not in users['user_id'].values:
            print("User not found.")
            return pd.DataFrame()
        current = start_point
        is_user = True  # Current node is a user

    # Pre-construct the graph as adjacency lists for better performance
    # This builds the graph structure only once for multiple walks
    user_to_movies = {}
    movie_to_users = {}

    # Build user -> movies adjacency list with weights
    for user_id in users['user_id'].unique():
        user_movies = ratings[ratings['user_id'] == user_id]
        user_movies = user_movies[user_movies['rating'] > 0]  # Only positive ratings

        if not user_movies.empty:
            # Store movies and their weights for this user
            user_to_movies[user_id] = {
                'movies': user_movies['movie_id'].values,
                'weights': user_movies['rating'].values.astype(float)
            }
            # Normalize weights
            if user_to_movies[user_id]['weights'].sum() > 0:
                user_to_movies[user_id]['weights'] /= user_to_movies[user_id]['weights'].sum()

    # Build movie -> users adjacency list with weights
    for movie_id in movies['movie_id'].unique():
        movie_users = ratings[ratings['movie_id'] == movie_id]
        movie_users = movie_users[movie_users['rating'] > 0]  # Only positive ratings

        if not movie_users.empty:
            # Store users and their weights for this movie
            movie_to_users[movie_id] = {
                'users': movie_users['user_id'].values,
                'weights': movie_users['rating'].values.astype(float)
            }
            # Normalize weights
            if movie_to_users[movie_id]['weights'].sum() > 0:
                movie_to_users[movie_id]['weights'] /= movie_to_users[movie_id]['weights'].sum()

    # Initialize tracking variables
    path = []
    visited_movies = set()

    # Perform random walk
    for _ in range(walk_length):
        if is_user:
            # Current node is a user, select a movie
            if current not in user_to_movies:
                break  # No movies for this user

            user_data = user_to_movies[current]
            if len(user_data['movies']) == 0:
                break

            # Select next movie based on weighted probability
            next_movie = np.random.choice(
                user_data['movies'],
                p=user_data['weights']
            )

            # Update tracking
            current = next_movie
            is_user = False

            # Only record movies in our path
            path.append(current)
            visited_movies.add(current)
        else:
            # Current node is a movie, select a user
            if current not in movie_to_users:
                break  # No users for this movie

            movie_data = movie_to_users[current]
            if len(movie_data['users']) == 0:
                break

            # Select next user based on weighted probability
            next_user = np.random.choice(
                movie_data['users'],
                p=movie_data['weights']
            )

            # Update tracking
            current = next_user
            is_user = True
            # Note: We don't add users to our path, only movies

    # Generate recommendations based on visit frequency
    if not path:
        return pd.DataFrame(columns=["Movie Name"])

    # Count movie visits and prepare recommendations
    movie_counts = pd.Series(path).value_counts()
    recommendations = pd.DataFrame({
        'movie_id': movie_counts.index,
        'count': movie_counts.values
    })

    # Add movie titles and format result
    recommendations = recommendations.merge(movies[['movie_id', 'title']], on='movie_id')
    recommendations = recommendations.sort_values('count', ascending=False)
    recommendations = recommendations[['title']].head(num).reset_index(drop=True)
    recommendations.index = recommendations.index + 1  # 1-based indexing
    recommendations = recommendations.rename(columns={'title': 'Movie Name'})

    return recommendations

# Example usage:
print("User-based recommendations:")
print(weighted_pixie_recommend(1, walk_length=15, num=5), "\n")

print("Movie-based recommendations:")
print(weighted_pixie_recommend("Jurassic Park (1993)", walk_length=10, num=5))
````
